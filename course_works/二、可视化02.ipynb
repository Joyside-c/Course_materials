```python
tt=list(map(float,data_u.可支配收入.tolist()))
uu = list(map(float,data_u.消费支出.tolist()))
```

```python
ttt=[]
for i in tt:
    t0=i/54305.3 *100
    ttt.append(t0)
uuu=[]
for i in uu:
    u0=i/37458.3*100
    uuu.append(u0)
```

```python
#第二个图用收入和支出作比较

r = (
    Geo()
    # 加载图表模型中的中国地图
    .add_schema(maptype="china")
    # 在地图中加入点的属性
    .add("可支配收入", [list(z) for z in zip(a1, ttt)])
    .add("消费支出", [list(z) for z in zip(a1, uuu)])
    # 设置坐标属性
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    # 设置全局属性
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(),
        #title_opts=opts.TitleOpts(title="Geo-基本示例"),
    )
)

# 在 html(浏览器) 中渲染图表
r.render()
# 在 Jupyter Notebook 中渲染图表
r.render_notebook()
```
